```python
Data1 = "/home/thanuja/Downloads/multi-lang emotion data/output_Greek/train/"
Data2 = "/home/thanuja/Downloads/multi-lang emotion data/output_Greek/test/"
```

```python
dir_train = os.listdir(Data1)
dir_test = os.listdir(Data2)

print(dir_train)
```

```python
emotion=[]
path=[]

for sub_dir in dir_train:
    filename = os.listdir(Data1+sub_dir)
    for files in filename:
        if files[0:1]=='a':
            emotion.append('thymomenos')
        elif files[0:1]=='d':
            emotion.append('aidia')
        elif files[0:1]=='f':
            emotion.append('fovos')
        elif files[0:1]=='h':
            emotion.append('charoumenos')
        elif files[0:1]=='s':
            emotion.append('lypimenos')
        else:
            emotion.append('Error')
        path.append(Data1+sub_dir+'/'+files)
```

```python
data_emo = pd.DataFrame(emotion,columns=['Emotion'])
data_emo.head()
```

```python
data_path = pd.DataFrame(path,columns=['Path'])
pd.set_option('display.max_colwidth',None)
data_path.head()
```

```python
train_data = pd.concat([data_path,data_emo],axis=1)
train_data.head()
```

```python
train_data['Emotion'].value_counts()
```

```python
speech_data=pd.read_csv("Greek_Data1.csv")
speech_data
```

```python
Feature_data = pd.DataFrame(columns=['Features'])

counter = 0
for index, path in enumerate(speech_data.Path):
    X, sample_rate = librosa.load(path,res_type='kaiser_fast')
    mfccs=np.mean(librosa.feature.mfcc(y=X, sr=sample_rate,n_mfcc=40).T, axis=0)
    Feature_data.loc[counter]=[mfccs]
    counter=counter+1
```

```python
Feature_data.head()
```

```python
Feature_data=pd.concat([pd.DataFrame(Feature_data['Features'].values.tolist()),speech_data.Emotion],axis=1)
Feature_data
```

```python
X_data = Feature_data.drop(['Emotion'], axis=1)
X_data
```

```python
Y_data = Feature_data.Emotion
Y_data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_data,Y_data,test_size=0.2,random_state=40)
```

```python
from sklearn.svm import SVC
svc_model=SVC(C=100,gamma=0.000001,kernel='rbf').fit(x_train,y_train)

print('accuracy: {}'.format(svc_model.score(x_test,y_test)))
```

```python
train_acc = float(svc_model.score(x_train, y_train)*100)
print("----train accuracy score %s ----" % train_acc)

test_acc = float(svc_model.score(x_test, y_test)*100)
print("----test accuracy score %s ----" % test_acc)
```

```python
from sklearn.svm import SVC
svc_model=SVC(C=100,kernel='linear').fit(x_train,y_train)

print('accuracy: {}'.format(svc_model.score(x_test,y_test)))
```

```python
train_acc = float(svc_model.score(x_train, y_train)*100)
print("----train accuracy score %s ----" % train_acc)

test_acc = float(svc_model.score(x_test, y_test)*100)
print("----test accuracy score %s ----" % test_acc)
```
